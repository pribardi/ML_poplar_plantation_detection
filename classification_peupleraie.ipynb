```python
import pandas as pd

# Charger le fichier CSV
csv_file_path = '3_tuiles_ready_b.csv'
data = pd.read_csv(csv_file_path)

# Afficher les premières lignes des données
print("Les premières lignes des données :")
print(data.head())

# Explorer les informations sur les données
print("\nInformations sur les données :")
print(data.info())

# Séparer les fonctionnalités (features) et la cible (target)
X = data.drop(columns=['Classe', 'ID', 'CODE_TFV', 'TFV', 'TFV_G11', 'ESSENCE', 'ID_plot', 'Dep', 'layer', 'path', 'originfid']) 
y = data['Classe']
```

```python
from sklearn.model_selection import KFold
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score
import numpy as np
X = X.values
y = y.values
# Définir le nombre de plis pour la validation croisée
num_folds = 10

# Initialiser la validation croisée KFold
kf = KFold(n_splits=num_folds, shuffle=True, random_state=7)

# Initialiser une liste pour stocker les modèles de chaque expérience de validation croisée
models = []

# Initialiser une variable pour suivre le meilleur score d'exactitude
best_accuracy = 0.0
best_model = None

# Boucler sur chaque expérience de validation croisée
for train_index, val_index in kf.split(X):
    X_train, X_val = X[train_index], X[val_index]
    y_train, y_val = y[train_index], y[val_index]

    # Initialiser un modèle de forêt aléatoire
    rf_model = RandomForestClassifier(n_estimators=100, random_state=7)

    # Entraîner le modèle sur les données d'entraînement
    rf_model.fit(X_train, y_train)

    # Prédire les étiquettes de validation
    y_pred = rf_model.predict(X_val)

    # Calculer le score d'exactitude pour ce modèle
    accuracy = accuracy_score(y_val, y_pred)

    # Sauvegarder le modèle et son score
    models.append((rf_model, accuracy))

    # Mettre à jour le meilleur modèle si nécessaire
    if accuracy > best_accuracy:
        best_accuracy = accuracy
        best_model = rf_model

# Sélectionner le meilleur modèle
print(f"Meilleur score d'exactitude : {best_accuracy:.2f}")
print("Caractéristiques du meilleur modèle : ", best_model)
```

# Grid Search = hyepr parameter finetuning + cross validation 

```python
from sklearn.model_selection import GridSearchCV

# Define the parameter grid for grid search
param_grid = {
    'n_estimators': range(10, 151, 10),  # Number of decision trees
    'max_depth': [None] + list(range(5, 51, 10)),  # Maximum depth of trees
    'max_features': ['auto'] + list(range(1, 21, 2))  # Number of features to consider for split
}

# Initialize the Random Forest Classifier
rf_model = RandomForestClassifier(random_state=7)

# Initialize GridSearchCV
grid_search = GridSearchCV(estimator=rf_model, param_grid=param_grid, cv=10, scoring='accuracy')

# Perform grid search
grid_search.fit(X, y)

# Get the best parameters and best model from grid search
best_params = grid_search.best_params_
best_model = grid_search.best_estimator_

# Print the best parameters
print("Best Parameters:", best_params)

# Retrain the best model on the entire training set
best_model.fit(X, y)

joblib.dump(best_model, 'meilleur_modele.pkl')
```
